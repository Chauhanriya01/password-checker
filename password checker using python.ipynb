PASSWORD CHECKER

```python
def is_strong_password(password):
    """this function chechs if the password is strong or weak"""
    if len(password)<8:
        return "the password is weak! add more characters to make it 8 characters long"
    if not any(char.isdigit() for char in password):
        return "Try a stronger one! Add a digit, lowercase, uppercase, and a special symbol."
    if not any(char.islower() for char in password):
        return "Try a stronger one! Add a lowercase letter, uppercase, and a special symbol."
    if not any(char.isupper() for char in password):
        return "Try a stronger one! Add an uppercase letter and a special symbol."
    if not any(char in '!@#$%^&*()_+' for char in password):
        return "Try a stronger one! Add a special character."
    return "this is a strong password"
        
        



```

```python
print(is_strong_password("yht_123"))
```

```python
print(is_strong_password("Y@mi123456"))
```
